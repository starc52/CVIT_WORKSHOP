### Function for distance of line from origin

```python
def distance_from_origin(a, b, c):
    return abs(c)/math.sqrt(a*a+b*b)
print(distance_from_origin(1.0, 0.0, -50.0))
print(distance_from_origin(0.0, 1.0, -20.0))
print(distance_from_origin(1.0, 0.0, -30.0))
```

### Function for distance between parallel lines

```python
def distance_parallel_lines(a, b, c1, c2):
    return abs(c1-c2)/math.sqrt(a*a+b*b)
print(distance_parallel_lines(1.0, 1.0, 9, 16))
print(distance_parallel_lines(1.0, -1.0, 14, 16))
```

### Ellipse

```python
# x*x/9+y*y/16 = 1
x = np.linspace(-3, 3, 50)
y1 = np.sqrt(16-(16/9)*np.power(x, 2))
y2 = -np.sqrt(16-(16/9)*np.power(x, 2))
%matplotlib qt
plt.plot(x, y1)
plt.plot(x, y2)
```

### Plotting $x+y+10=0$ in 3D

```python
from mpl_toolkits.mplot3d import Axes3D
a = 1
b = 1
c = 0
d = 10
xx = np.linspace(-5, 5, 20)
zz = np.linspace(-5, 5, 20)
X, Z = np.meshgrid(xx, zz)
yy = (-a*X - c*Z - d) * 1.0 / b
%matplotlib qt
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, yy, Z)
```

### Plot of Gaussian Function

```python
xx = np.linspace(-5, 5, 200)
yy = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(xx, yy)
zz = np.exp(-X*X-Y*Y)
%matplotlib qt
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, Y, zz)
```
